Earthquake Data

```python
import pandas as pd

df = pd.read_csv('earthquake_data.csv', 
    header=0, 
    names=[ 'Worried about earthquakes', 
            'Worried about the "Big One"?', 
            'Will the "Big One" occur in your lifetime', 
            'Experienced an earthquake', 
            'Taken any precautions for an earthquake', 
            'Familiar with the San Andreas Fault line', 
            'Familiar with the Yellowstone Supervolcano', 
            'Age', 
            'Gender', 
            'Household earning last year', 
            'US Region'])
df
```

```python
age_gender = df[['Age', 'Gender', 'Will the "Big One" occur in your lifetime']]
age_gender
```

```python
age_count = age_gender.groupby(['Will the "Big One" occur in your lifetime', 'Age']).count()
age_count
```

```python
gender_count = age_gender.groupby(['Will the "Big One" occur in your lifetime', 'Gender']).count()
gender_count
```
